```python
#  FUNCTION TO GET ALL TABLE DATA INTO A JSON FILE
import html
import json
def get_all_columns_to_file(table_data, file_name):
    '''
    Create a json file with all columns from a table data
    :param table_data: table data
    :param file_name: file name
    '''
    columns_list = []
    for row in table_data:
        for column in row.keys():
            if column not in columns_list:
                columns_list.append(column)

    with open(file_name, 'w', encoding="utf-8") as f:
        json.dump(columns_list, f, indent=4, ensure_ascii=False)
```

```python
def get_Data(path):
    with open(path, 'r', encoding="utf8") as f:
        return json.load(f)
    
def exportData(dataBase, path:str):
    with open(path, 'w', encoding="utf-8") as f:
        json.dump(dataBase, f, indent=4, ensure_ascii=False)

def createDict(database, key, value, path):
    data = {}
    for item in database:
        objectKey = item.get(key)
        objectValue = item.get(value)

        # Verifique se a chave já existe no dicionário antes de adicioná-la
        if objectKey not in data:
            data[objectKey] = objectValue
    exportData(data, path)

def getDataConverted(dataBase, index: int ,key: str, subkey: str, columnFlow):
    valueKey = str(dataBase[index].get(key, {}).get(subkey, None))
    if valueKey in columnFlow:
        return (columnFlow[valueKey])
    else:
        return (None)
    
def converterDataFromDCIdToFlowId(dealcloud_id:int, from_to:dict[str,int]) ->int|None:
    if str(dealcloud_id) in from_to:
        return (from_to[str(dealcloud_id)])
    return None
                

def getFirstDataInColumn(dataBase, index:int, column, subkey: int) -> int:
    column_values = dataBase[index].get(column)
    if column_values is None:
        return None
    if isinstance(column_values,list):
        for value in column_values:
            subkey_value = value.get(subkey)
            if subkey_value is not None:
                return subkey_value
        return None
    else:
        return column_values.get(subkey)


def getDataInColumnAndConvertToFlow(dataBase, index, column, subkey, colunaFlow):
    originalData = getFirstDataInColumn(dataBase, index, column, subkey)
    dataConvertedToFlow = converterDataFromDCIdToFlowId(str(originalData), colunaFlow)
    return dataConvertedToFlow

# companiesDealCloud = loadDataDC('./data-dealcloud/company.json')
```
